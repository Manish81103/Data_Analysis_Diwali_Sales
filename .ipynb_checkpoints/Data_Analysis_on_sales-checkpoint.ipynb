```python
# import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_csv("C:\\Users\\manish singh\\OneDrive\\projects\\Diwali_Sales_Data\\Diwali Sales Data.csv", encoding= 'unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status','unnamed1','User_ID'],axis=1,inplace=True)
```

```python
#check for null values
round(df.isnull().mean()*100,2)
```

```python
#drop nal values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount']=df['Amount'].astype('int')
```

```python
df.columns
```

```python
df.describe()
```

```python
df.head()
```

```python
#Filling Unmarried and Married in place of 0 and 1 repectively

dic={0:"Unmarried",1:"Married"}
print(dic)
```

```python
df['Marital_Status']=df['Marital_Status'].map(dic)
```

```python
df.head()
```

# <center>==Graphical Representation Of Data==<center>

# <center>On The Basis Of Gender<center>

```python
val=df['Gender'].value_counts().values
print(val)
```

```python
ind=df['Gender'].value_counts().index
print(ind)
```

```python
# Percentage of male and female
color=['pink','lightskyblue']
plt.pie(x=val,labels=ind,autopct='%1.2f%%',colors=color)
plt.title('Gender Distribution',size=20)
plt.show()
```

```python
# plotting a bar chart for Gender and it's count

ax=sns.countplot(df,x='Gender',palette=color)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Count Of Gender',size=20)
plt.xlabel('Gender',fontsize=15)
plt.ylabel('count',fontsize=15)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
sales=[74335853,31913276]
```

```python
# Sales amount by gender
sns.barplot(sales_gen,x='Gender',y='Amount',palette=color)
for index, data in enumerate(sales):
    plt.text(x=index,y=data+29,s=f'{data}',ha='center')

plt.title('Sales Amount By Gender',size=20)
plt.xlabel('Gender',fontsize=15)
plt.ylabel('Amount',fontsize=15)
```

> From the above graphs we can see that most of the customers are females and even the purchasing power of females are greater   than males and female customers has spent the more amount than males.

# <center>On The Basis Of Age<center>

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sales_age=sales_age.sort_index()
```

```python
sales_age
```

```python
sorted=sales_age['Age Group'].sort_index()
```

```python
# count of different age group by gender

plt.figure(figsize=(10,5))
ax=sns.countplot(df,x='Age Group',hue='Gender',order=sorted,palette=color)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title(' Age Group With Gender',size=20)
plt.xlabel('Age Group',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.show()
```

```python
# Sales Amount of  different Age Group
plt.figure(figsize=(8,6))
ax=sns.barplot(sales_age,x='Age Group',y='Amount',palette='Pastel1')
  
for p in ax.patches:ax.annotate(p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),  
                    ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')


plt.title('Sales Amount By Age Group',size=20)  
plt.xlabel('Age Group',fontsize=15)
plt.ylabel('Amount',fontsize=15)
plt.show()
```

 > From the graphs we can see that most of the customers are from 26-35 age group and least from 0-17

# <center>On The Basis Of State<center>

```python
# Total number of orders from top 10 states

sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

plt.figure(figsize=(16.3,5))
ax=sns.barplot(sales_state,x='State',y='Orders',palette='Pastel1')
for bar in ax.containers:
    ax.bar_label(bar)

plt.title('Number Of Orders From Top 10 States',size=20)
plt.xlabel('State',fontsize=15)
plt.ylabel('Orders',fontsize=15)
plt.show()

    
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
sales_state
```

```python
state_amount=[19374968,14427543,13523540,11603818,8101142,8037146,4963368,4220175,4022757,3946082]
```

```python
# Sale amount from top 10 states
plt.figure(figsize=(17,5))
sns.barplot(sales_state,x='State',y='Amount',palette='Pastel1')
for index, data in enumerate(state_amount):
    plt.text(x=index,y=data+29,s=f'{data}',ha='center')

plt.title('Sales Amount From Top 10 States',size=20)
plt.xlabel('State',fontsize=15)
plt.ylabel('Amount',fontsize=15)
plt.show()
```

> From the above graphs we can see that Uttar Pradesh has the highest orders with highest sales amount.Where as Kerala, Haryana   and Gujarat has the lowest orders but Haryana,Bihar and Gujarat has the lowest sales amount.

#  <center>On The Basis of Product Category<center>

```python
# Count of different products

plt.figure(figsize=(10,4))
ax=sns.countplot(df,x='Product_Category',palette='Pastel1')
for bar in ax.containers:
    ax.bar_label(bar)
plt.xticks(rotation=90)
plt.title('Count Of Product Category',size=20)
plt.xlabel('Product Category',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()    
```

```python
# Top 5 most selling products
sales_pr=df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)

plt.figure(figsize=(10,4))
ax=sns.barplot(sales_pr,x='Product_Category',y='Orders',palette='Pastel1')
for bar in ax.containers:
    ax.bar_label(bar)
plt.title('Top 5 Most Selling Products',size=20)
plt.xlabel('Product Category',fontsize=15)
plt.ylabel('Orders',fontsize=15)
plt.show()
```

```python
sales_pro=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sales_pro
```

```python
pro_amount=[33933883,16495019,15643846,15575209,5440051]
```

```python
# Sales amount of  top 5 products

plt.figure(figsize=(10,4))
sns.barplot(sales_pro,x='Product_Category',y='Amount',palette='Pastel1')
for index, data in enumerate(pro_amount):
    plt.text(x=index,y=data+29,s=f'{data}',ha='center')


plt.title('Sales Amount Of Top 5 Products',size=20)
plt.xlabel('Product Category',fontsize=15)
plt.ylabel('Amount',fontsize=15)
plt.show()
```

From the above graphs we can see that Clothing & Apparel are the most selling products but Food items has highest sales amount
